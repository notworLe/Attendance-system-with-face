```python
import cv2
from pathlib import Path
from retinaface import RetinaFace
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('data/img/Screenshot (28).png', cv2.IMREAD_COLOR_RGB)

plt.imshow(img)
plt.show()
```

```python
faces = RetinaFace.detect_faces(img)
print(faces)
```

```python
a = {
    '1': 1,
    '2': 2,
    '3': 3
}
for i in a:
    print(i)
```

```python
for key, val in faces.items():
    x1, y1, x2, y2 = val['facial_area']
    cv2.rectangle(img, (x1, y1), (x2, y2), color=(41, 245, 214), thickness=2)
```
